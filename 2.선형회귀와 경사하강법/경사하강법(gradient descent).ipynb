```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([1, 4.5, 9, 10, 13]) # data X 좌표
y = np.array([0, 0.2,2.5,5.4,7.3]) # data Y 좌표

w, b = 0, 0
learning_rate, epoch = 0.005, 1000 # 보폭, 최대 걸음 수(반복횟수)
n = len(X)
```

```python
for i in range(epoch): # epoch 만큼 반복
    y_pred = w * X + b # 현재 (w, b)지점에서의 y 추정 값
    error = y_pred - y # 그 떄의 error
    w = w - learning_rate * (error * X).sum()
    b = b - learning_rate * error.sum()

print('w =', w.round(2), ', b =',b.round(2))
error
```

```python
plt.scatter(X, y, color='blue', marker='D')

# 계산으로 구한 w, b를 이용하여 선형 회귀 직선을 그리자
y_pred = w * X - b
plt.plot(X, y_pred, 'r:')
plt.show()
```
