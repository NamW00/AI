```python
import numpy as np
import matplotlib.pylab as plt
```

```python
def step_func(x):
    return 1.0*(x>0)
```

```python
def sigmoid_func(x):
    return 1/(1+np.exp(-x))
```

```python
def relu_func(x):
    return np.maximum(x, 0)
```

```python
x = np.arange(-20,20, 0.1)
plt.plot(x, step_func(x), label='step')
plt.plot(x, sigmoid_func(x), linestyle='--', label='sigmoid')
plt.legend()
plt.show()
```

```python
plt.plot(x, np.tanh(x), label='tanh')

plt.legend()
plt.show()
```

```python
plt.plot(x, relu_func(x), linestyle='--', label='relu')

plt.legend()
plt.show()
```
