```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-7,7,0.1) # [0, 0.1, 0.2, ..., 5.9]
a = np.exp(x)
b = np.exp(-x)
y1 = (1/(1+b))
y2 = ((a-b)/(a+b))
```

```python
# plot은 해당 점들을 연속으로 찍어주면서 표현 
plt.plot(x, y1, label = 'Function 1')
plt.plot(x, y2, linestyle = '--',label = 'Function 2')
plt.xlabel('x')  # x축 이름
plt.ylabel('f(x)')  # y축 이름
plt.title('Assignment1')  # 제목
plt.legend() # 범례
plt.show()
```
