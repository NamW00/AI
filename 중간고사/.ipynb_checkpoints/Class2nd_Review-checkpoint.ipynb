1. 평균 제곱 오차(Mean squared error,MSE)

```python
import numpy as np
```

```python
y = np.array([1, 2, 3, 4, 5]) # 실제 데이터
y_hat = np.array([1.2, 2.4, 2.5, 4.6, 5.4]) # 추정치

diff = (y - y_hat) ** 2
print("오차 : ", y - y_hat) # 오차
print("오차들의 제곱 : ", diff)
e_mse = np.average(diff)
print('평균 제곱오차 =', e_mse)
```

2. 경사하강법(Gradient descent)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([1, 4.5, 9, 10, 13])
Y = np.array([0, 0.2, 2.5, 5.4, 7.3])

w, b = 0, 0
learning_rate, epoch = 0.005, 1000 # 보폭, 최대 걸음수(반복 횟수)
n = len(X)
```

```python
for i in range(epoch): # 학습 루프
    y_pred = w*X + b   # 현재 w, b를 이용한 작업 T
    error = y_pred - y # 성능 척도 P
    w = w - learning_rate * (error*X).sum() # 경험 E로 개선
    b = b - learning_rate * error.sum()

print('w = ', w.round(2), ', b =', b.round(2))
error
```

```python
plt.scatter(X, y, color='blue', marker='D')

y_pred = w*X -b
plt.plot(X, y_pred, 'r:')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([-3, -2, -1, 0, 1, 2, 3])
y = np.array([13.13, 8.16, 2.83, 1.33, 0.58, 2.53, 7.42])

w2, w1, b = 0, 0, 0
learning_rate, epoch = 0.005, 1000
n = len(x)
```

```python
for i in range(epoch):
    y_pred = w2*x**2 + w1*x + b
    error = y_pred - y
    w2 = w2 - learning_rate * (error * x**2).sum()
    w1 = w1 - learning_rate * (error * x).sum()
    b = b - learning_rate * error.sum()
print('w2 =', w2.round(2), 'w1 =', w1.round(2), 'b = ', b.round(2))
error
```

```python
plt.scatter(x, y, color='blue', marker='D')

y_pred = w2*x**2 + w1*x + b
plt.plot(x, y_pred, 'r:')
plt.show()
```
